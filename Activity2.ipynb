```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('diabetes_model.csv')
```

# 1 Creat count plot of outcome

```python
sns.countplot(data['Outcome'])
```

# 2 Analyse the effect of age on outme using boxplot

```python
sns.boxplot(x='Outcome',y='Age',data=data)
plt.title('Effect of age on outcome')
```

# 3 Check for chances of getting diabetes if it is common in family, using the Diabetes Pedigree Function

```python
sns.violinplot(x=data['Outcome'],y=data['DiabetesPedigreeFunction'],data=data)
plt.title('chances getting diabetes ')
```

# 4. Check for the relationship with Insulin & Glucose and group the data points by Outcome. 

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x=data['Insulin'],y=data['Glucose'],hue=data['Outcome'],data=data)
```

# 5 Create a heatmap on Correlation of features in the data.

```python
sns.heatmap(data.corr(),annot=True)
```
